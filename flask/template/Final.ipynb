# Data Preprocessing

```python
data.drop('Case_No',axis=1,inplace=True)
```

```python
# Segregating the categorical and numerical columns into different variables for ease of understanding
cat_cols=data.select_dtypes(include='object')
num_cols=data.select_dtypes(exclude='object')
```

```python
cat_cols
```

```python
num_cols
```

```python
# filling all the missing values in 'Age Months' column with median
data['Age Months'].fillna(data['Age Months'].median(),inplace=True)
data['Qchat Score'].fillna(data['Qchat Score'].median(),inplace=True)
```

```python
# filling missing values in coulmns using ffill method
a=['A1','A2','A3','A4','A5']
for i in a:
    data[i].ffill(axis=0,inplace=True)
    
# filling missing values in coulmns using bfill method
b=['A6','A7','A8','A9','A10']
for i in b:
    data[i].bfill(axis=0,inplace=True)
```

```python
# To find out the correlation values of features
corr=data.corr()
corr
```

```python
categorical_cols=['Sex','Ethnicity','Jaundice','Family mem with ASD','Who completed the test']
encoder=OneHotEncoder(drop='first',sparse=False)
encoder_cols=pd.DataFrame(encoder.fit_transform(data[categorical_cols]),columns=encoder.get_feature_names_out(categorical_cols))

numerical_cols=['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','Age Months']
scaler=StandardScaler()
scaled_cols=pd.DataFrame(scaler.fit_transform(data[numerical_cols]),columns=scaler.get_feature_names_out(numerical_cols))
```

# Model Building

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,matthews_corrcoef,roc_curve,roc_auc_score,confusion_matrix
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.utils.fixes import loguniform
```

```python
data.replace({'ASD Traits':{'Yes':1,'No':0}},inplace=True)
```

```python
X=pd.concat([encoder_cols,scaled_cols],axis=1)
Y=data['ASD Traits']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
log=LogisticRegression()
log.fit(X_train,Y_train)
```

```python
pred_train = log.predict(X_train)
pred_test = log.predict(X_test)
```

```python
print(classification_report(Y_test,pred_test))
```

```python
roc=log.predict_proba(X_train)[:,1]
fpr, tpr, threshold = roc_curve(Y_train,roc)
plt.plot([0,1], [0,1],'k--') #x=(0,1) y
plt.plot(fpr,tpr,label='logistic')
plt.ylabel("TPR")
plt.xlabel("FPR")
plt.show()
```

```python
roc_auc_score(Y_train,roc)
```

```python
mcc= matthews_corrcoef(Y_test,pred_test)
print("MCC: ",mcc)
```

```python
cm = confusion_matrix(Y_test,pred_test)
print('Confusion Matrix: ',cm)
sns.heatmap(cm, annot=True, fmt = '.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# Hyperparameter Tuning 

# Grid search

```python
#grid search tuning parameters are deinfed, using gridsearchcv model is trained
param_grid ={
    'penalty':["l1","l2"],
    'C': [0.1,0.5,1,5,10]
} #l1 lasso l2 ridge to regularize the model , C common factor alpha LOS
grid=GridSearchCV(estimator=log, param_grid=param_grid, cv=5)
grid.fit(X_train,Y_train)
```

```python
# best paramters and best model is chosen 
best_param = grid.best_params_
best_model = grid.best_estimator_
# predicted values of y are generated based on best model on the test data
y_pred_t=best_model.predict(X_test)
```

```python
#Machine choses the best parameters 
print('Best hyperparameters: ',best_param)
```

```python
#the classification report is generated on the testing values
print(classification_report(Y_test, y_pred_t))
```

```python
cm = confusion_matrix(Y_test,y_pred_t)
print('Confusion Matrix: ',cm)
sns.heatmap(cm, annot=True, fmt = '.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# Randomized Search

```python
#randomized search tuning parameters are deinfed, using randomized searchcv the model is trained
logistic_random_param={'C':loguniform(1e-4,1e0),
                      'max_iter':(np.arange(100,800,10))}
grid=RandomizedSearchCV(estimator=log, param_distributions=logistic_random_param, cv=5)
grid.fit(X_train,Y_train)
```

```python
# best paramters and best model is chosen, predicted values of y are generated based on best model on the test data
best_param = grid.best_params_
best_model = grid.best_estimator_
y_pred_tr=best_model.predict(X_test)
```

```python
#Machine choses the best parameters 
print("Best Hyperparameters:", best_param)
```

```python
#the classification report is generated on the testing values
print(classification_report(Y_test, y_pred_tr))
```

```python
cm = confusion_matrix(Y_test,y_pred_tr)
print('Confusion Matrix: ',cm)
sns.heatmap(cm, annot=True, fmt = '.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
